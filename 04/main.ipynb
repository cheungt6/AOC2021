```python
class bingo_card(object):
    def __init__(self, num_array: [], x: int, y: int):
        self.num_array = num_array
        self.numbers_remaining = set()
        self.lines = [] # set()
        self.x = x
        self.y = y
        
        for i in range(self.y):
            col = set()
            for j in range(self.x):
                num = num_array[i][j]
                self.numbers_remaining.add(num)
                col.add(num_array[j][i])
            self.lines.append(set(num_array[i])) # row
            self.lines.append(col)
            
    def remove(self, i: int):
        if i in self.numbers_remaining:
            #print(f"found number {i}")
            #print(self.lines)
            self.numbers_remaining.remove(i)
            for line in self.lines:
                if i in line:
                    line.remove(i)
                    if len(line) == 0:
                        #print(self.numbers_remaining)
                        return True
        return False
```

```python
cards = []
numbers = []
x = 5
y = 5
with open('input', 'r') as f:
    line = f.readline()
    numbers = [int(i) for i in line.split(",")]
    line = f.readline()
    
    while line:
        num_array = []
        for i in range(y):
            line = f.readline()
            row = [int(i) for i in line.split(" ") if i != ""]
            num_array.append(row)
        #print(num_array)
        card = bingo_card(num_array, x, y)
        cards.append(card)
        line = f.readline()
```

```python
print(len(cards))
print(numbers)
completed = []
def run() -> int:
    for i in numbers:
        for idx in range(len(cards)):
            card = cards[idx]
            if card.remove(i):
                print(f"{i} x {(sum(card.numbers_remaining))} = {(i * sum(card.numbers_remaining))}")
                completed.append(card)
                print(len(cards))
        for r in completed:
            if r in cards:
                cards.remove(r)
            
run()
```
